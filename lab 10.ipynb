```python
import numpy as np

# Feature vector (state-action features)
x = np.array([2, 3])

# Weights
w = np.array([0.5, 1.0])

# Q-value approximation
Q = np.dot(w, x)

print("Approximated Q-value:", Q)
```

```python
alpha = 0.1
gamma = 0.9

# current state features
x = np.array([2, 3])

# next state features
x_next = np.array([1, 2])

w = np.array([0.5, 1.0])

# reward
r = 5

Q = np.dot(w, x)
Q_next = np.dot(w, x_next)

# TD error
delta = r + gamma * Q_next - Q

# update weights
w = w + alpha * delta * x

print("Updated weights:", w)
```
