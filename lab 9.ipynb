```python
import numpy as np

n_states = 5
n_actions = 2
n_options = 2

# Q-values
Q_high = np.zeros((n_states, n_options))   
Q_low = np.zeros((n_states, n_actions))    

alpha = 0.1
gamma = 0.9

def step(state, action):
    next_state = (state + action + 1) % n_states
    reward = 1 if next_state == 4 else 0
    return next_state, reward

# Option policies (simple mapping)
def choose_option(state):
    return np.argmax(Q_high[state])

def choose_action(state):
    return np.argmax(Q_low[state])

# Training
for episode in range(200):
    state = np.random.randint(n_states)

    for _ in range(20):
        option = choose_option(state)
        action = choose_action(state)

        next_state, reward = step(state, action)

        # Low-level update
        Q_low[state, action] += alpha * (
            reward + gamma * np.max(Q_low[next_state]) - Q_low[state, action]
        )

        # High-level update
        Q_high[state, option] += alpha * (
            reward + gamma * np.max(Q_high[next_state]) - Q_high[state, option]
        )

        state = next_state

print("High-level Q:\n", Q_high)
print("Low-level Q:\n", Q_low)
```
