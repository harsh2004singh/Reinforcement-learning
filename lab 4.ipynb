```python
import numpy as np

# Simple MDP (3 states)
states = 3
gamma = 0.9

# Reward matrix [state][action]
R = np.array([
    [5, 10],
    [3, 2],
    [0, 0]
])

# Transition (deterministic for simplicity)
T = np.array([
    [1, 2],  # from state 0
    [2, 0],  # from state 1
    [2, 2]   # terminal
])

V = np.zeros(states)

for _ in range(100):  # iterations
    new_V = np.zeros(states)
    for s in range(states):
        values = []
        for a in range(2):
            next_s = T[s][a]
            values.append(R[s][a] + gamma * V[next_s])
        new_V[s] = max(values)
    V = new_V

print("Optimal State Values (DP):", V)
```

```python
import numpy as np

states = 3
gamma = 0.9
alpha = 0.1

V = np.zeros(states)

# Simulated transitions (s, r, s')
episodes = [
    (0, 5, 1),
    (1, 3, 2),
    (0, 10, 2),
    (1, 2, 0)
]

for _ in range(100):
    for (s, r, s_next) in episodes:
        V[s] += alpha * (r + gamma * V[s_next] - V[s])

print("State Values (TD):", V)
```
