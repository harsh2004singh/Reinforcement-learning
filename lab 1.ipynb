```python
import numpy as np

# Environment
class Bandit:
    def __init__(self, k=5):   
        self.k = k
        self.means = np.random.rand(k)

    def pull(self, arm):
        return np.random.randn() + self.means[arm]
```

```python
# UCB Algorithm
def ucb_bandit(bandit, steps=1000):
    k = bandit.k
    counts = np.zeros(k)
    values = np.zeros(k)

    for i in range(k):
        reward = bandit.pull(i)
        counts[i] += 1
        values[i] = reward

    for t in range(k, steps):
        ucb = values + np.sqrt(2 * np.log(t + 1) / counts)
        arm = np.argmax(ucb)
        reward = bandit.pull(arm)

        counts[arm] += 1
        values[arm] += (reward - values[arm]) / counts[arm]

    return values
```

```python
# PAC (ε-Greedy)
def epsilon_greedy(bandit, epsilon=0.1, steps=1000):
    k = bandit.k
    counts = np.zeros(k)
    values = np.zeros(k)

    for _ in range(steps):
        if np.random.rand() < epsilon:
            arm = np.random.randint(k)
        else:
            arm = np.argmax(values)

        reward = bandit.pull(arm)
        counts[arm] += 1
        values[arm] += (reward - values[arm]) / counts[arm]

    return values 
```

```python
# Run
bandit = Bandit()
print("UCB Values:", ucb_bandit(bandit))
print("PAC ε-Greedy Values:", epsilon_greedy(bandit))
```
