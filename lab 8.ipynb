```python
import numpy as np

# States: 0 = clean, 1 = dirty
states = [0, 1]

# Belief (initial)
belief = np.array([0.6, 0.4])

# Transition matrix P(s'|s,a)
# Action: clean (0), wait (1)
T = {
    0: np.array([[1, 0], [1, 0]]),   # clean → always clean
    1: np.array([[0.7, 0.3], [0.2, 0.8]])  # wait
}

# Observation model P(o|s)
# o=0 clean observed, o=1 dirty observed
O = np.array([
    [0.9, 0.1],  # state clean
    [0.2, 0.8]   # state dirty
])

gamma = 0.9

def belief_update(belief, action, observation):
    new_belief = np.zeros(len(states))

    for s_prime in states:
        prob = 0
        for s in states:
            prob += T[action][s][s_prime] * belief[s]
        new_belief[s_prime] = O[s_prime][observation] * prob

    # Normalize
    new_belief /= np.sum(new_belief)
    return new_belief

# Simulation
action = 1        # wait
observation = 1   # observed dirty

new_belief = belief_update(belief, action, observation)

print("Old Belief:", belief)
print("Updated Belief:", new_belief)
```
